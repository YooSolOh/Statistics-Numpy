```python
# 25도 이상=1, 아니면 0
hot_days = (df_temp['temp'] >= 25)
print(hot_days.unique())

hot_days = (df_temp['temp'] >= 25).astype(int)
print(hot_days.unique())

hot_days = (df_temp['temp'] >= 25).astype(int).values
print(hot_days)
```

```python
summer_df = []

# 연속 5일 체크
for i in range(len(hot_days) - 4):
    if np.sum(hot_days[i:i+5]) == 5:
        print(f"\n여름 시작일: {df_temp.loc[i, ['Date', 'temp']]}")
        summer_st = df_temp.loc[i, ['Date', 'radiation_short','radiation_long']].to_dict()
        summer_df.append(summer_st)

# DataFrame으로 만들기
summer_start_df = pd.DataFrame(summer_df)

# CSV로 저장하기
summer_start_df.to_csv('summer_start_dates.csv', index=False)
```
